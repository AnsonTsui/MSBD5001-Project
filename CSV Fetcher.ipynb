# Fetcher

```python
import requests
from bs4 import BeautifulSoup
import csv
from time import sleep

# Send a GET request to the website and retrieve the HTML content for each year
base_url = "https://hoopshype.com/salaries/players/"
years = range(1990, 2024)
sum_entr = 0

# Prepare the CSV file
filename = "nba_player_salaries.csv"
with open(filename, "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Year", "Player Name", "Salary (Unadjusted)", "Salary (Adjusted)"])

    # Iterate over the years
    for year in years:

        if year == 2023:
            url = base_url
        else:
            url = f"{base_url}{year}-{year+1}/"

        response = requests.get(url)
        html_content = response.text

        # Parse the HTML content using Beautiful Soup
        soup = BeautifulSoup(html_content, "html.parser")

        # Find the table containing player salaries
        table = soup.find("table", class_="hh-salaries-ranking-table")

        # Extract player names and salaries from the table
        data = []
        for row in table.find_all("tr"):
            cells = row.find_all("td")
            if len(cells) >= 3:
                if cells[1].text.strip() != "Player":
                    player_name = cells[1].text.strip()
                    salary_unadjusted = cells[2].text.strip()

                    if year == 2023:
                        salary_adjusted = salary_unadjusted
                    else:
                        salary_adjusted = cells[3].text.strip()

                    # Append the data for the current year
                    #print([f"{year}-{year+1}", player_name, salary_unadjusted, salary_adjusted])
                    data.append([f"{year}-{year+1}", player_name, salary_unadjusted, salary_adjusted])

        # Write the data for the current year to the CSV file
        print(f"{year}-{year+1} : {len(data)} entries")
        sum_entr += len(data)
        writer.writerows(data)

print(f"Data saved successfully to {filename}.")
print(f"{sum_entr} in total")
```

```python
def read_csv_file(filename):
    """Reads a CSV file and returns its contents as a list of rows."""
    with open(filename, "r") as csvfile:
        reader = csv.reader(csvfile)
        rows = list(reader)
    return rows

def compare_csv_files(file1, file2):
    """Compares two CSV files and returns the different rows."""
    csv1 = read_csv_file(file1)
    csv2 = read_csv_file(file2)

    # Extract the identifier column index
    identifier_index = 0  # Assuming the first column is the identifier

    # Create sets of identifiers for quick comparison
    identifiers1 = set(row[identifier_index] for row in csv1)
    identifiers2 = set(row[identifier_index] for row in csv2)

    # Find the different and missing rows
    different_rows = [row for row in csv1 if row[identifier_index] not in identifiers2]
    different_rows += [row for row in csv2 if row[identifier_index] not in identifiers1]

    return different_rows

# Specify the filenames of the CSV files
csv_file1 = "/content/nba_player_salaries.csv"
csv_file2 = "/content/result (2).csv"

# Compare the CSV files
different_rows = compare_csv_files(csv_file1, csv_file2)

# Print the different rows
for row in different_rows:
    print(row)
```
